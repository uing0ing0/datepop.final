```python
try:
    datepop_data = pd.read_csv(
        "data/datepop/shop_seoul_crawled.csv", header=0, encoding='cp949')
    datepop_data.reset_index(drop=True, inplace=True)
except Exception as e:
    print("파일 불러오기 실패:", e)
```

```python
# NaN 값을 0으로 채울 컬럼들
columns_to_zero = ['instagram_post', 'instagram_follower',
                   'visitor_review_count', 'blog_review_count', 'distance_from_subway']
# NaN 값을 False로 채울 컬럼들
columns_to_false = ['on_tv', 'seoul_michelin', 'on_blue_ribbon', "no_kids",
                    "parking_available", "hot_spot", "age-2030", "gender-balance", "new_store"]

# NaN 값이 있는 경우 해당 값으로 채워넣기
for column in columns_to_zero:
    datepop_data[column].fillna("0", inplace=True)

for column in columns_to_false:
    datepop_data[column].fillna(False, inplace=True)
```

```python
datepop_data.columns
```

```python
category_data = {
    "food": [],
    "alcohol": [],
    "cafe": [],
    "plyaing": [],
    "place_lend": [],
    "care": [],
    "movie": [],
    "else": [],
}

def categorize(row, category_list):
    if row['category'] in category_list:
        return True
    else:
        return False

category_data['food'] = datepop_data[datepop_data.apply(categorize, category_list=food, axis=1)]
category_data['alcohol'] = datepop_data[datepop_data.apply(categorize, category_list=alcohol, axis=1)]
category_data['cafe'] = datepop_data[datepop_data.apply(categorize, category_list=cafe, axis=1)]
category_data['playing'] = datepop_data[datepop_data.apply(categorize, category_list=playing, axis=1)]
category_data['place_lend'] = datepop_data[datepop_data.apply(categorize, category_list=place_lend, axis=1)]
category_data['care'] = datepop_data[datepop_data.apply(categorize, category_list=care, axis=1)]
category_data['movie'] = datepop_data[datepop_data.apply(categorize, category_list=movie, axis=1)]
category_data['else'] = datepop_data[datepop_data.apply(categorize, category_list=else_list, axis=1)]
```

### Food

```python
food = category_data["food"]
food.reset_index(drop=True, inplace=True)
```

```python
store_scores = pd.DataFrame(columns=["Naver ID", "Name", "Score"])

for index, item in food.iterrows():
    score = calculate_score(item)
    print(f"ID: {item["store_id"]} 매장 이름: {item["name"]} 매장 점수: {score}")

    print("="*50)
    result = pd.DataFrame({"Naver ID": [item["store_id"]], "Name": [item["name"]], "Score": [score]})
    store_scores = pd.concat([store_scores, result], ignore_index=True)
    
```

```python
store_scores
```

```python
store_scores = store_scores.sort_values(by="Score", ascending=False, ignore_index=True).head(20)
store_scores
```

```python
store_scores.to_csv('data/datepop/shop_score.csv', encoding='utf-8-sig')
```
